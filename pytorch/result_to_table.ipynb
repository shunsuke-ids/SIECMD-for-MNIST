```python
import pandas as pd
pd.options.display.float_format = '{:.3f}'.format
```

```python
df = pd.read_csv('resnet18_cfv8_results.csv', sep=',')
df.head()
```

```python
groups = df.groupby('loss')[['best_acc', 'best_cmae', 'final_acc', 'final_cmae']].agg(['mean', 'std'])
groups
```

```python
groups.xs('mean', axis=1, level=1)
```

```python
groups.style.format('{:.3f}').highlight_max(axis=0, subset=[('best_acc', 'mean'), ('final_acc', 'mean')], color='lightgreen').highlight_min(axis=0, subset=[('best_cmae', 'mean'), ('final_cmae', 'mean')], color='lightcoral')
```
